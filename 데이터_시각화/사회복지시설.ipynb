```python
swf.columns = ["시도", "시군구", "계", "국가", "지자체", "사회복지법인", "비영리법인"]


swf["행정구역"] = swf["시도"] + " " + swf["시군구"]
swf[["행정구역", "계", "국가", "지자체", "사회복지법인", "비영리법인"]]
```

```python
swf = swf.iloc[2:].reset_index(drop=True)


#swf.columns = ["시도", "시군구", "계", "국가", "지자체", "사회복지법인", "비영리법인"]



swf = swf[swf["시군구"] != "소계"].copy()


swf["행정구역"] = swf["시도"] + " " + swf["시군구"]


result_sum = swf[["행정구역", "계"]]
result_nation = swf[["행정구역","국가"]]
result_local_government = swf[["행정구역","지자체"]]
result_swfoundation = swf[["행정구역","사회복지법인"]]
result_non_profit = swf[["행정구역","비영리법인"]]
# 결과 확인
print(result_sum)
```

```python
result_sum.loc[:, "계"] = pd.to_numeric(result_sum["계"])
result_sum.loc[:, "계비율"] = result_sum["계"] / 1278
result_sum_rate = result_sum.drop(columns=["계"])
result_sum_rate_sorted = result_sum_rate.sort_values(by="계비율", ascending=False)
result_sum_rate_sorted
```

```python
result_nation.loc[:, "국가"] = pd.to_numeric(result_nation["국가"])
result_nation.loc[:, "국가 비율"] = result_nation["국가"] / 1278
result_nation_rate = result_nation.drop(columns=["국가"])
result_nation_rate_sorted = result_nation_rate.sort_values(by="국가 비율", ascending=False)
result_nation_rate_sorted
```

```python
result_local_government.loc[:, "지자체"] = pd.to_numeric(result_local_government["지자체"])
result_local_government.loc[:, "지자체 비율"] = result_local_government["지자체"] / 958
result_local_government.info()

result_local_government_rate = result_local_government.drop(columns=["지자체"])
result_local_government_rate_sorted = result_local_government_rate.sort_values(by="지자체 비율", ascending=False)
result_local_government_rate_sorted
```

```python
result_swfoundation.loc[:, "사회복지법인"] = pd.to_numeric(result_swfoundation["사회복지법인"])
result_swfoundation.loc[:, "사회복지법인 비율"] = result_swfoundation["사회복지법인"] / 958
result_swfoundation.info()

result_swfoundation_rate = result_swfoundation.drop(columns=["사회복지법인"])
result_swfoundation_rate_sorted=result_swfoundation_rate.sort_values(by="사회복지법인 비율", ascending=False)
result_swfoundation_rate_sorted
```

```python
result_non_profit.loc[:, "비영리법인"] = pd.to_numeric(result_non_profit["비영리법인"])
result_non_profit.loc[:, "비영리법인 비율"] = result_non_profit["비영리법인"] / 958
result_swfoundation.info()

result_non_profit_rate = result_non_profit.drop(columns=["비영리법인"])
result_non_profit_rate_sorted = result_non_profit_rate.sort_values(by="비영리법인 비율",ascending=False)
result_non_profit_rate_sorted

```
